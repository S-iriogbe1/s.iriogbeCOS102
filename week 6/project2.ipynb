```python
import tkinter as tk

class ShippingManager:
    def __init__(self, destination, weight):
        self.destination = destination.lower()
        self.weight = weight

    def get_price(self):
        # Shipping rates based on destination
        if self.destination == "local":
            return self.weight * 5
        elif self.destination == "national":
            return self.weight * 10
        elif self.destination == "international":
            return self.weight * 25
        else:
            return None

# Button logic
def calculate_shipping():
    dest = dest_entry.get().strip()
    try:
        weight = float(mass_entry.get())
        if weight <= 0:
            output_label.config(text="Weight must be greater than 0", fg="red")
            return
    except ValueError:
        output_label.config(text="Enter a valid number for weight", fg="red")
        return

    manager = ShippingManager(dest, weight)
    price = manager.get_price()

    if price is not None:
        output_label.config(text=f"Total Shipping Price: ${price}", fg="green")
    else:
        output_label.config(text="Location not supported (local/national/intl)", fg="red")

# GUI Setup
app = tk.Tk()
app.title("Shipping Cost Checker")
app.geometry("400x350")

tk.Label(app, text="Enter Destination:", font=("Arial", 10)).pack(pady=5)
dest_entry = tk.Entry(app)
dest_entry.pack(pady=5)

tk.Label(app, text="Enter Weight (kg):", font=("Arial", 10)).pack(pady=5)
mass_entry = tk.Entry(app)
mass_entry.pack(pady=5)

calc_btn = tk.Button(app, text="Calculate Price", command=calculate_shipping, bg="green", fg="white")
calc_btn.pack(pady=20)

output_label = tk.Label(app, text="", font=("Arial", 11, "bold"))
output_label.pack(pady=10)

app.mainloop()
```
