```python
import random
from tkinter import *

class StaffManager:
    def __init__(self):
        self.staff_list = {
            "John": "Manager", "Alice": "Developer", "Bob": "HR",
            "Charlie": "Support", "David": "Marketing", "Eve": "Sales",
            "Frank": "Design", "Grace": "Finance", "Heidi": "Legal",
            "Ivan": "IT", "Judy": "Security", "Karl": "Operations",
            "Linda": "Admin", "Mike": "Engineering", "Nancy": "Customer Service"
        }
        self.staff_pool = ["John", "Alice", "Bob", "Charlie", "David"]
        self.tasks = ["Reviewing reports", "Answering emails", "Coding features", "Client meeting", "Delivering items"]
        self.present_staff = []

    def is_valid_staff(self, name):
        return name in [s.lower() for s in self.staff_list]

    def is_staff_present(self, name):
        return name.lower() in [s.lower() for s in self.present_staff]

    def give_task(self, name):
        return f"Task: {random.choice(self.tasks)} assigned to {name}"

    def confirm_ready(self):
        return "System ready for check-in."

    def deny_entry(self):
        return "ACCESS DENIED: Staff not registered."

# Logic for button
def handle_click():
    user_input = name_entry.get().strip()
    if staff_manager.is_valid_staff(user_input):
        output_label.config(text=f"Welcome {user_input}!\n{staff_manager.give_task(user_input)}", fg="green")
    else:
        output_label.config(text=staff_manager.deny_entry(), fg="red")

# Initialize StaffManager
staff_manager = StaffManager()

# Tkinter Setup
app = Tk()
app.title("Staff Check-in System")
app.geometry("400x300")

title_label = Label(app, text="ENTER STAFF NAME", font=("Arial", 14, "bold"))
title_label.pack(pady=10)

name_entry = Entry(app, font=("Arial", 12))
name_entry.pack(pady=5)

check_button = Button(app, text="Check-in", command=handle_click, font=("Arial", 12), bg="blue", fg="white")
check_button.pack(pady=10)

output_label = Label(app, text="", font=("Arial", 10))
output_label.pack(pady=20)

app.mainloop()
```
