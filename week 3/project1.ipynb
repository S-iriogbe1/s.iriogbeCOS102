```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd

# 1. Load the dataset
try:
    # Ensure this file exists in the same folder as your script
    data = pd.read_csv("employee_verification.csv")
except FileNotFoundError:
    # We use a hidden root for the error if the main window isn't ready
    messagebox.showerror("Error", "CSV file 'employee_verification.csv' not found!")
    exit()

# 2. Function to check employee credentials
def check_employee():
    name = name_entry.get().strip()
    dept = dept_entry.get().strip()

    # Fixed syntax: Python uses 'not', not '!'
    if not name or not dept:
        messagebox.showwarning("Input Error", "Please enter both name and department")
        return
    
    # Convert to lowercase for case-insensitive matching
    name_lower = name.lower()
    dept_lower = dept.lower()

    # Filter data: Looks for a row where both Name and Dept match
    found_employee = data[(data['Name'].str.lower() == name_lower) & 
                          (data['Department'].str.lower() == dept_lower)]

    if not found_employee.empty:
        # If found, show welcome message and list colleagues
        dept_members = data[data['Department'].str.lower() == dept_lower]
        members_list = ", ".join(dept_members['Name'].tolist())
        
        result_text.set(f"Welcome, {name}!\n\nOther members in {dept}:\n{members_list}")
    else:
        # If no match found
        result_text.set("")
        messagebox.showinfo("Result", "Employee does not exist in this department.")

# 3. Create GUI Layout
root = tk.Tk()
root.title("Logistics Employee Verification")
root.geometry("400x400")

# Input Fields
tk.Label(root, text="Enter Name:", font=("Arial", 10, "bold")).pack(pady=5)
name_entry = tk.Entry(root, width=30)
name_entry.pack(pady=5)

tk.Label(root, text="Enter Department:", font=("Arial", 10, "bold")).pack(pady=5)
dept_entry = tk.Entry(root, width=30)
dept_entry.pack(pady=5)

# Action Button
check_button = tk.Button(root, text="Check Employee", command=check_employee, bg="#4CAF50", fg="white")
check_button.pack(pady=20)

# Output display area
result_text = tk.StringVar()
result_label = tk.Label(root, textvariable=result_text, wraplength=350, justify="left", fg="blue")
result_label.pack(pady=10)

# Start the app
root.mainloop()
```
