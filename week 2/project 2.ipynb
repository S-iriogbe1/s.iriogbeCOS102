```python
import numpy as np

def solve_quadratic():
    print("\n Quadratic Equation: Ax^2 + Bx + C = 0 ")
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    roots = np.roots([a, b, c])
    print(f"Roots: {roots}")

def solve_cubic():
    print("\n Cubic Equation: Ax^3 + Bx^2 + Cx + D = 0 ")
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))
    roots = np.roots([a, b, c, d])
    print(f"Roots: {roots}")

def solve_quartic():
    print("\n Quartic Equation: Ax^4 + Bx^3 + Cx^2 + Dx + E = 0 ")
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))
    e = float(input("Enter E: "))
    roots = np.roots([a, b, c, d, e])
    print(f"Roots: {roots}") 

def main():
    while True:
        print("\n Polynomial Root Finder ")
        print("1. Solve Quadratic (Degree 2)")
        print("2. Solve Cubic (Degree 3)")
        print("3. Solve Quartic (Degree 4)")
        print("4. Exit")

        choice = input("\nSelect an operation (1-4): ") 

    
        if choice == '1':
            solve_quadratic()
        elif choice == '2':
            solve_cubic()
        elif choice == '3':
            solve_quartic()
        elif choice == '4':
            print("Mission accomplished. Goodbye!")
            break  
        else:
            print("Invalid selection. Please try again.")

if __name__ == "__main__":
    main()

    
    
```
