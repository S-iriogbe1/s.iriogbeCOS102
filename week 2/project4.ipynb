```python
experience = int(input("Enter your years of experience: "))
age = int(input("Enter your age: "))

if experience > 25 and age >= 55:
    print("Your annual tax resume is $5,600,000")
elif experience > 20 and age >= 45:
    print("Your annual tax revenue is $4,480,000")
elif experience > 10 and age >= 35:
    print("Your annual tax revenue is $1,500,000")
elif experience < 10 and age < 35:
    print("Your annual tax revenue is $550,000")
else:
    print("fall to find category, try again later, sorry..")
```
